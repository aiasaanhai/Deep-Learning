```python
#import torch package

import torch
```

```python
# check if NVIDIA CUDA is available or not

torch.cuda.is_available()
```

```python
# GPU device name

torch.cuda.get_device_name()
```

```python
#creating a tensor

torch.tensor([2, 3, 6, 9, 10])
```

```python
#Creating a tensor with random values from uniform distribution on interval [0, 1)
torch.rand(2, 3)
```

```python
#Creating a tensor with random values from standard normal distribution
torch.randn(3, 2)
```

```python
# Creating zero-valued tensor
torch.zeros(2, 3)
```

```python
#Creating tensor with ones
torch.ones(2, 3)
```

```python
# Transform/Reshape tensors

a = torch.rand(3, 4)
print(a)
a = a.reshape(6, 2)
print(a)
```

```python
# the rank of the tensor is increased to 4. There are 3 channels, where each row has 4 columns
a = torch.rand(3, 4)
a.reshape(3, 1, 4)
```

```python
# unsqueeze adds a dimension of length one at a specified position

a = torch.rand(3)
print(a)
print(torch.unsqueeze(a,1))
```

```python
# Matrix multiplication

a = torch.rand(2, 4)
b = torch.rand(4, 3)

print(a)
print("====================================")
print(b)
print("====================================")
print(torch.mm(a, b))
```

```python
# Element-wise multiplication of two tensors

a = torch.tensor([[[ 0., 1., 2.],
                   [ 3., 4., 5.]]])
print(a)
print("====================================")
b = torch.rand(1, 3)
print(b)
print("====================================")
print(torch.mul(a, b))
```

```python
# Element-wise multiplication of tensor with a scaler

a = torch.rand(2, 3)
print(a)
print("====================================")
b = 100
print(torch.mul(a, b))
```

```python
# Corss product of two matrix

a = torch.tensor([[0. , 1. , 2.],
                 [3. , 4. , 5.]])

print(a)
print("====================================")

b = torch.tensor([[6., 7. , 8.],
                  [9., 10. , 11.]])
print(b)
print("====================================")
torch.cross(a, b)
```
